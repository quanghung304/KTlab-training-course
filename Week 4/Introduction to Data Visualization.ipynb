```python
import pandas as pd
iris = pd.read_csv('iris.csv', index_col = 0)
iris.head()
```

```python
wine_reviews = pd.read_csv('archive/winemag-data-130k-v2.csv', index_col=0)
wine_reviews.head()
```

```python
#I/ Matplotlib
#1. Scatter Plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots() #need to declare a figure to put an axis on it
ax.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
#We can give the graph more meaning by coloring in each data-point by its class
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
def recode_species(Species):
    if Species == 'Iris-setosa':
        return 0
    if Species == 'Iris-versicolor':
        return 1
    else:
        return 2
iris['species_color'] = iris['Species'].apply(recode_species)
ax.scatter(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'], c=iris['species_color'])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
#2. Line chart
# get columns to plot
columns = iris.columns.drop(['Species', 'species_color'])
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

```python
iris = iris.drop(['species_color'], axis = 1)
iris
```

```python
#3. Histogram
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(wine_reviews['points'])
# set title and labels
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
#4. Bar chart
# count the occurrence of each class 
data = wine_reviews['points'].value_counts() 
data
```

```python
fig, ax = plt.subplots() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Wine Review Scores') 
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

```python
#2 Heatmap
#for Matplotlib 
import numpy as np

# get correlation matrix
corr = iris.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the x tick labels and set their alignment. ha means horizontal alignment {'center', 'right', 'left'}
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

#To add annotations to the heatmap we need to add two for loops:
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       #va means vertical alignment {'center', 'top', 'bottom', 'baseline', 'center_baseline'}
                       ha="center", va="center", color="black")
```

```python
#for Seaborn, we just need 1 line :)
sns.heatmap(iris.corr(), annot=True)
```

```python
#3 Faceting
#Faceting is the act of breaking data variables up across multiple subplots and combining those subplots into a single figure.
#First of all, we need to define the FacetGrid and pass it our data as well as a row or column, which will be used to split the data
g = sns.FacetGrid(iris, col='Species')
#Then we need to call the map function on our FacetGrid object and define the plot type we want to use, as well as the column we want to graph.
g = g.map(sns.kdeplot, 'SepalLengthCm')
```
