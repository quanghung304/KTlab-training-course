```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
fig = plt.figure()
axe = fig.add_subplot(1,1,1)
axe.hist(tips['total_bill'], bins = 10)
axe.set_title('Histogram of Total bill')
axe.set_xlabel('Frequency')
axe.set_ylabel('Total Bill')
fig
```

```python
scatter_plot = plt.figure()
axe = scatter_plot.add_subplot(1,1,1)
axe.scatter(tips['total_bill'], tips['tip'])
axe.set_xlabel('Total Bill')
axe.set_ylabel('Tips')
axe.set_title('Scatterplot of Total Bill and Tip')
scatter_plot
```

```python
boxplot = plt.figure()
axe = boxplot.add_subplot(1,1,1)
axe.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
    tips [tips ['sex'] == 'Male']['tip']],
    labels = ['Female', 'Male'])
axe.set_xlabel('Sex')
axe.set_ylabel('Tip')
axe.set_title('Boxplot of Tips by Sex')
boxplot
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else: 
        return 1
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()
```

```python
scatter_plot = plt.figure()
axe = scatter_plot.add_subplot(1,1,1)
axe.scatter(x = tips['total_bill'], y = tips['tip'],
           s = tips['size'] * 10, c = tips['sex_color'])
axe.set_xlabel('Total Bill')
axe.set_ylabel('Tips')
axe.set_title('Total bill and Tip colored by Sex and sized by Size')
scatter_plot
```
